```python
# Import Spark and create a SparkSession
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("BigData-HW-1").config("spark.driver.extraClassPath","/content/postgresql-42.2.9.jar").getOrCreate()
```

# Extract the Amazon Data into Spark DataFrame

```python
# Read in the data from an S3 Bucket
from pyspark import SparkFiles
url= "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Mobile_Electronics_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
mobile_electronics_review_data_df= spark.read.csv(SparkFiles.get("amazon_reviews_us_Mobile_Electronics_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)
mobile_electronics_review_data_df.show()
```

```python
# Get the number of rows in the DataFrame.
rows = mobile_electronics_review_data_df.count()
print(f"DataFrame Rows count : {rows}")
```

# Transform the Data

## Create the "review_id_table".

```python
from pyspark.sql.functions import to_date
# Create the "review_id_df" DataFrame with the appropriate columns and data types.
review_id_df = mobile_electronics_review_data_df.select(["review_id","customer_id","product_id","product_parent",
                                          to_date("review_date","yyyy-MM-dd").alias("review_date")])
review_id_df.show()
```

## Create the "products" Table

```python
# Create the "products_df" DataFrame that drops the duplicates in the "product_id" and "product_title columns. 
products_df = mobile_electronics_review_data_df.select(["product_id","product_title"]).drop_duplicates()

products_df.show()
```

## Create the "customers" Table

```python
# Create the "customers_df" DataFrame that groups the data on the "customer_id" by the number of times a customer reviewed a product. 
from pyspark.sql.functions import sum,avg,max,count
customers_df = mobile_electronics_review_data_df.groupby('customer_id').agg(count("*").alias("customer_count"))
customers_df.show()
```

## Create the "vine_table".

```python
# Create the "vine_df" DataFrame that has the "review_id", "star_rating", "helpful_votes", "total_votes", and "vine" columns. 
vine_df = mobile_electronics_review_data_df.select(["review_id","star_rating","helpful_votes","total_votes","vine"])
vine_df.show()
```

# Load

```python
mode = "append"
jdbc_url="jdbc:postgresql://endpoint"
config = {"user":"postgres", "password": "<password>", "driver":"org.postgresql.Driver"}
```

```python
# Write review_id_df to table in RDS
review_id_df.write.jdbc(url=jdbc_url, table='review_id_table', mode=mode, properties=config)
```

```python
# Write products_df to table in RDS
products_df.write.jdbc(url=jdbc_url, table='products', mode=mode, properties=config)
```

```python
# Write customers_df to table in RDS
customers_df .write.jdbc(url=jdbc_url, table='customers', mode=mode, properties=config)
```
